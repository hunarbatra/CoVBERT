```python
import torch
from transformers import RobertaTokenizer, RobertaModel
from bertviz import head_view, model_view
import re
```

```python
def call_html():
  import IPython
  res = IPython.core.display.HTML('''
        <script src="/static/components/requirejs/require.js"></script>
        <script>
          requirejs.config({
            paths: {
              base: '/static/base',
              "d3": "https://cdnjs.cloudflare.com/ajax/libs/d3/3.5.8/d3.min",
              jquery: '//ajax.googleapis.com/ajax/libs/jquery/2.0.0/jquery.min',
            },
          });
        </script>
        ''')
  print(res)
  html = res.data
  with open('test.html', 'w') as f:
      f.write(html)
```

```python
model = RobertaModel.from_pretrained("hunarbatra/CoVBERT", output_attentions=True)
tokenizer = RobertaTokenizer.from_pretrained("hunarbatra/CoVBERT")
```

```python
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
```

```python
model = model.to(device)
model = model.eval()
```

```python
def show_head_view(model, tokenizer, sequence):
    inputs = tokenizer.encode_plus(sequence, return_tensors='pt', add_special_tokens=True)
    input_ids = inputs['input_ids']
    attention = model(input_ids.to(device))[-1]
    input_id_list = input_ids[0].tolist() # Batch index 0
    tokens = tokenizer.convert_ids_to_tokens(input_id_list)    
    head_view(attention, tokens)
```

```python
def show_model_view(model, tokenizer, sequence):
    inputs = tokenizer.encode_plus(sequence, return_tensors='pt', add_special_tokens=True)
    input_ids = inputs['input_ids']
    attention = model(input_ids.to(device))[-1]
    input_id_list = input_ids[0].tolist() # Batch index 0
    tokens = tokenizer.convert_ids_to_tokens(input_id_list)    
    model_view(attention, tokens)
```
